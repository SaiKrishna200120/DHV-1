# Visualization of obesity data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ObesityDataSet_raw_and_data_sinthetic.csv')
df
```

Creating new column called BMI

```python
df['BMI']=df['Weight']/df['Height']**2
```

```python
df.describe()
```

Grouping the data


```python
data =df.groupby('Age')
groupage = data.mean('Age')
groupage=groupage.reset_index()
groupage
```

```python
data1 = df.groupby('Gender')
groupgen = data1.mean('Gender')
groupgen = groupgen.reset_index()
```

```python
sns.barplot(x='Gender',y='CH2O',data=groupgen)
plt.yscale('log')
```

```python
sns.barplot(x='Gender',y='FCVC',data=groupgen)
plt.yscale('log')
```

```python
sns.scatterplot(x='Age',y='BMI',data =groupage,hue=df['Gender'])
```

```python
sns.scatterplot(x='Age',y='Height',data =groupage,hue=df['Gender'])
```

```python
sns.scatterplot(x='Age',y='Weight',data =groupage,hue=df['Gender'])
```

```python
 sns.lmplot(x='Age', y='BMI', hue='Gender', data=df, fit_reg=True)
```

```python
 sns.lmplot(x='Age', y='Height', hue='Gender', data=df, fit_reg=True)
```

```python
 sns.lmplot(x='Age', y='Weight', hue='Gender', data=df, fit_reg=True)
```

```python
sns.histplot(x='Age',data=df,bins=10)
```

```python
sns.histplot(x='Height',data=df,bins=10)
```

```python
sns.histplot(x='Weight',data=df,bins=10)
```

```python
sns.histplot(x='BMI',data=df,bins=10)
```

```python
sns.heatmap(df[['Age','Height','Weight','FCVC','NCP','CH2O','FAF','TUE','BMI']].corr(),annot=True)
```

```python
sns.boxplot(df)
plt.grid()
plt.yscale('log')
```
